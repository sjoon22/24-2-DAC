```python
from PIL import Image

def crop_and_resize(img, output_size=(256, 256)):

    width, height = img.size

    # 정방형으로 자르기
    if width > height:
        left = (width - height) // 2
        top = 0
        right = left + height
        bottom = height
    else:
        top = (height - width) // 2
        left = 0
        bottom = top + width
        right = width

    # 크롭
    img = img.crop((left, top, right, bottom))

    # 크기 변경
    img = img.resize(output_size)

    return img
```

```python
import numpy as np

def estimate_keypoints(image, openpose):
        x = np.uint8(image.convert('RGB'))
        keypoints = openpose(x)
        keypoints = keypoints[0] if len(keypoints) > 0 else np.zeros((18, 3), dtype=np.int32)
        keypoints[np.where(keypoints[:, 2]==0), :2] = -1
        keypoints = keypoints[:, :2]
        return keypoints

def keypoints2heatmaps(keypoints, size=256):
        heatmaps = np.zeros((size, size, keypoints.shape[0]), dtype=np.float32)
        for k in range(keypoints.shape[0]):
            x, y = keypoints[k]
            if x == -1 or y == -1:
                continue
            heatmaps[y, x, k] = 1.0
        return heatmaps
```

```python
import pandas as pd

keypoints_columns = ['file_id', 'img_h', 'img_w', 'p0_x', 'p0_y', 'p1_x', 'p1_y', 'p2_x', 'p2_y', 'p3_x', 'p3_y', 'p4_x', 'p4_y',
 'p5_x', 'p5_y', 'p6_x', 'p6_y', 'p7_x', 'p7_y', 'p8_x', 'p8_y', 'p9_x', 'p9_y',
 'p10_x', 'p10_y', 'p11_x', 'p11_y', 'p12_x', 'p12_y', 'p13_x', 'p13_y', 'p14_x',
 'p14_y', 'p15_x', 'p15_y', 'p16_x', 'p16_y', 'p17_x', 'p17_y', 'filepath']
base_path = "/content/drive/MyDrive/24-2/데캡디/"
folders = ["M2", "M3", "M4"]
person = ["P01", "P02", "P03", "P04", "P05", "P06", "P07", "P08", "P09", "P10"]
```

```python
import os
from openpose.body.estimator import BodyPoseEstimator

save_path = os.path.join(base_path, "pose2pose/dataset/DeepFashion/")
openpose = BodyPoseEstimator(pretrained=True)
keypoints_csv_path = os.path.join(save_path, "keypoints.csv")
pairs_csv_path = os.path.join(save_path, "pairs.csv")
img_list_csv_path = os.path.join(save_path, "img_list.csv")

# pd.DataFrame(columns=keypoints_columns).to_csv(keypoints_csv_path, index=False)
# pd.DataFrame(columns=['imgA', 'imgB']).to_csv(pairs_csv_path, index=False)
# pd.DataFrame(columns=['img']).to_csv(img_list_csv_path, index=False)
```

```python
import os
from PIL import Image
import itertools

image_paths = []
pairs = []
keypoints = []
heatmaps = []



for main_folder in folders:
  main_path = os.path.join(base_path, main_folder)
  for p in person:
    path = os.path.join(main_path, p)
    for sub_folder in os.listdir(path):
      sub_path = os.path.join(path, sub_folder)
      print(sub_path)
      sub_image_paths = []
      for inner_folder in os.listdir(sub_path):
        inner_path = os.path.join(sub_path, inner_folder)
        img = Image.open(inner_path)
        img = crop_and_resize(img)
        id = "img"+inner_folder[:-4]
        temp = "img/" + inner_folder

        keypoint = estimate_keypoints(img, openpose)
        keypoints.append([id, 256, 256]+list(keypoint.flatten())+[temp])
        heatmap = keypoints2heatmaps(keypoint)
        np.savez(save_path+"/pose_maps/"+id+".npz", heatmap)

        img_path = os.path.join(os.path.join(save_path, "img"), inner_folder)
        img.save(img_path)


        sub_image_paths.append("img/" +inner_folder)
        image_paths.append("img/" +inner_folder)
      sub_path_pairs = list(itertools.combinations(sub_image_paths, 2))
      pairs.extend(sub_path_pairs)

      existing_data = pd.read_csv(keypoints_csv_path)
      new_data = pd.DataFrame(keypoints, columns=keypoints_columns)
      updated_data = pd.concat([existing_data, new_data], ignore_index=True)
      updated_data.to_csv(keypoints_csv_path, index=False)
      keypoints.clear()

      existing_pairs = pd.read_csv(pairs_csv_path)
      new_pairs = pd.DataFrame(sub_path_pairs, columns=["imgA", "imgB"])
      updated_pairs = pd.concat([existing_pairs, new_pairs], ignore_index=True)
      updated_pairs.to_csv(pairs_csv_path, index=False)

      existing_img_list = pd.read_csv(img_list_csv_path)
      new_img_row = pd.DataFrame({
        "img": [",".join(image_paths)]
      })
      updated_img_list = pd.concat([existing_img_list, new_img_row], ignore_index=True)
      updated_img_list.to_csv(img_list_csv_path, index=False)
      image_paths.clear()
```

```python
import pandas as pd

# Load the uploaded files
img_list_path = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/cleaned_img_list.csv'
keypoints_path = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/keypoints.csv'
pairs_path = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/pairs.csv'

# Read the data
img_list = pd.read_csv(img_list_path)
keypoints = pd.read_csv(keypoints_path)
pairs = pd.read_csv(pairs_path)

# Display first few rows to understand the structure
img_list.head(), keypoints.head(), pairs.head()
```

```python
# Extract the group identifier from the img filenames in img_list
img_list['group'] = img_list['img'].apply(lambda x: x.split('_')[2])

# Group by the extracted identifier and split into 8:2
from sklearn.model_selection import train_test_split

# Group data by 'group' identifier
grouped = img_list.groupby('group')

# Create a list of unique groups
unique_groups = grouped['img'].apply(list).tolist()

# Split the groups into train and test sets (8:2 ratio)
train_groups, test_groups = train_test_split(unique_groups, test_size=0.2, random_state=42)

# Flatten the nested lists for train and test
train_img_list = [img for group in train_groups for img in group]
test_img_list = [img for group in test_groups for img in group]

# Create DataFrames for train and test
train_img_df = pd.DataFrame({'img': train_img_list})
test_img_df = pd.DataFrame({'img': test_img_list})

# Verify the split
train_img_df.head(), test_img_df.head(), len(train_img_df), len(test_img_df)
```

```python
# Filter keypoints based on train and test image lists
train_keypoints = keypoints[keypoints['filepath'].isin(train_img_df['img'])]
test_keypoints = keypoints[keypoints['filepath'].isin(test_img_df['img'])]

# Filter pairs based on train and test image lists
train_pairs = pairs[
    pairs['imgA'].isin(train_img_df['img']) & pairs['imgB'].isin(train_img_df['img'])
]
test_pairs = pairs[
    pairs['imgA'].isin(test_img_df['img']) & pairs['imgB'].isin(test_img_df['img'])
]

# Verify the filtered data
train_keypoints.head(), test_keypoints.head(), train_pairs.head(), test_pairs.head()
```

```python
# Save the split data to separate CSV files
train_img_df.to_csv('/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/train_img_list.csv', index=False)
test_img_df.to_csv('/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/test_img_list.csv', index=False)
train_keypoints.to_csv('/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/train_img_keypoints.csv', index=False)
test_keypoints.to_csv('/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/test_img_keypoints.csv', index=False)
train_pairs.to_csv('/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/train_img_pairs.csv', index=False)
test_pairs.to_csv('/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/test_img_pairs.csv', index=False)
```

```python
import os
import pandas as pd
import shutil

# Paths to the files and directories
test_list_path = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/test_img_list.csv'
train_list_path = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/train_img_list.csv'
pose_maps_dir = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/pose_maps'
train_pose_maps_dir = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/train_pose_maps'
test_pose_maps_dir = '/content/drive/MyDrive/24-2/데캡디/pose2pose/dataset/DeepFashion/test_pose_maps'

# Load the file lists
def process_file_list(file_path):
    file_list = pd.read_csv(file_path, header=None)[0].tolist()
    print(file_list)
    return [f.replace(".jpg", ".npz").replace("/", "") for f in file_list]

test_list = process_file_list(test_list_path)
train_list = process_file_list(train_list_path)
print(test_list)
# Ensure target directories exist
os.makedirs(train_pose_maps_dir, exist_ok=True)
os.makedirs(test_pose_maps_dir, exist_ok=True)

# Function to move files
def move_files(file_list, target_dir):
    for file_name in file_list:
        source_path = os.path.join(pose_maps_dir, file_name)
        target_path = os.path.join(target_dir, file_name)
        if os.path.exists(source_path):
            shutil.move(source_path, target_path)

# Move the files based on lists
move_files(train_list, train_pose_maps_dir)
move_files(test_list, test_pose_maps_dir)

print(f"Train files moved: {len(os.listdir(train_pose_maps_dir))}")
print(f"Test files moved: {len(os.listdir(test_pose_maps_dir))}")
```

```python
test_pose_maps_files = os.listdir(test_pose_maps_dir)

# Find matched and unmatched files
moved_files = [f for f in test_list if f in test_pose_maps_files]
missing_files = [f for f in test_list if f not in test_pose_maps_files]

# Print summary
print(f"Total test files listed: {len(test_list)}")
print(f"Files successfully moved: {len(moved_files)}")
print(f"Missing files: {len(missing_files)}")

if missing_files:
    print("\nMissing files:")
    for file in missing_files:
        print(file)
```

jpg / png 등등은 상관 없음

size에 큰 영향을 받나 싶어서, 기존에 있던 condition 사진과 비슷한 사이즈로 resize 한 다음에 넣어봤는데 동일함.

1. 단색 다소 간단한 외형에 대해서만 작동을 잘함
2. 외국인 얼굴로 학습이 되어서 그런 것인지, 내 얼굴은 잘 안들어감 (빛의 차이도 있을 것 같음)

```python
from api import Pose2Pose
from PIL import Image
import matplotlib.pyplot as plt

# Pose2Pose 모델 초기화 (사전 학습된 모델 사용)
p2p = Pose2Pose(pretrained=True)

# 조건 이미지 및 참고 포즈 이미지 로드
condition = Image.open('./api-test/condition.jpg')  # 조건 이미지
reference = Image.open('./api-test/target_pose_reference.jpg')  # 참고 포즈

# 포즈 변환 수행
generated = p2p.transfer_as(condition, reference)

# 세 이미지 출력
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(condition)
ax[0].set_title('Condition Image')
ax[0].axis('off')

ax[1].imshow(reference)
ax[1].set_title('Reference Pose Image')
ax[1].axis('off')

ax[2].imshow(generated)
ax[2].set_title('Generated Pose Image')
ax[2].axis('off')

plt.show()
```
